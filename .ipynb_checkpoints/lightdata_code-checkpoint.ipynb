```python
import pandas as pd
import csv
```

```python
#url for 2023
url2023 = "https://aa.usno.navy.mil/calculated/durdaydark?year=2023&task=1&lat=38.89&lon=-77.03&label=Washington%2C+DC&tz=5&tz_sign=-1&submit=Get+Data"
data2023 = pd.read_html(url2023,header=1)[0]  # Assuming the table is the first one on the page
```

```python
# Create a DataFrame (example)
data2023 = pd.DataFrame(data2023)

# Add a new column 'Year' with the constant value '2023'
data2023 ['Year'] = 2023

# Display the updated DataFrame
data2023 

#re-arrage columns 
desired_columns = ['Day','Year','Jan.','Feb.', 'Mar.', 'Apr.', 'May', 'June', 'July', 'Aug.', 'Sep.', 'Oct.', 'Nov.', 'Dec.'] 
data2023 = data2023[desired_columns]

data2023 = data2023.reindex(columns=desired_columns)
data2023
```

```python
# Define the columns you want to keep as identifiers
id_vars = ['Day']

# Melt the DataFrame to convert columns into rows
data2023_melted = data2023.melt(id_vars=id_vars, var_name='Month', value_name='Daylight_Hours')

# Reorder the columns if needed
data2023_melted = data2023_melted[['Day', 'Month', 'Daylight_Hours']]

# Display the melted DataFrame
print(data2023_melted)
```

```python
# Create a new column combining the day with a fixed month and year
data2023['Full_date'] = '1/' + data2023['Day'].astype(str) + '/2023'

# Convert the "full_date" column to a datetime format
data2023['Full_date'] = pd.to_datetime(data2023['Full_date'], format='%m/%d/%Y')

data2023['Full_date'] = data2023['Full_date'].dt.strftime('%d/%m/%Y')

# Drop the original "day" column if needed
data2023 = data2023.drop(columns=['Day'])
```

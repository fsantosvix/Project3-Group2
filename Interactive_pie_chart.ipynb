### Use this code to filter the original dataframe and keep only the relevant data

```python
import pandas as pd
from pathlib import Path
```

```python
# Read data from csv file
street_lights_data = Path("Original_DBs/Street_Lights.csv")
df = pd.read_csv(street_lights_data)
df.head()
```

```python
# List all columns
df.columns
```

```python
# Create a filtered version with only the relevant columns
reduced_df = df[['X', 'Y', 'LIGHTTYPE_DESC', 'NUMBERLIGHTS','STREETLIGHTID', 'STREETNAME', 'WATTAGE1']]

reduced_df.head(5)
```

```python
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Group the DataFrame by street name and light type, and aggregate the number of lights
grouped_df = reduced_df.groupby(['STREETNAME', 'LIGHTTYPE_DESC'])['NUMBERLIGHTS'].sum().reset_index()

# Define unique street names
street_names = reduced_df['STREETNAME'].unique()

# Define colors for the pie chart using default color palette from Matplotlib
colors = plt.cm.tab10.colors 

def plot_pie_chart(street_name, chart_size):
    if street_name == 'All':
        aggregated_df = grouped_df.groupby('LIGHTTYPE_DESC')['NUMBERLIGHTS'].sum().reset_index()
        labels = aggregated_df['LIGHTTYPE_DESC']
        sizes = aggregated_df['NUMBERLIGHTS']
        title = 'Number of Lights by Light Type - All Streets'
    else:
        filtered_df = grouped_df[grouped_df['STREETNAME'] == street_name]
        labels = filtered_df['LIGHTTYPE_DESC']
        sizes = filtered_df['NUMBERLIGHTS']
        title = f'Number of Lights by Light Type - {street_name}'
    
    plt.figure(figsize=(chart_size, chart_size))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.show()

# Modify the dropdown options to include an option for 'All Streets'
all_options = ['All'] + list(street_names)
interact(plot_pie_chart, street_name=widgets.Dropdown(options=all_options, description='Select Street Name:'), chart_size=(4, 12, 1))
```

```python
interact(plot_pie_chart, street_name=widgets.Dropdown(options=street_names, description='Select Street Name:'), chart_size=(4, 12, 1))
```

```python
# Calculate the total wattage and add as a new column
reduced_df["Total_Wattage"] = reduced_df['NUMBERLIGHTS']*reduced_df['WATTAGE1']
reduced_df
```

```python
# Remove incomplete records
clean_df = reduced_df.dropna(how="any")
```

```python
# Export df as a csv file
clean_df.to_csv('filteredDF.csv', index=False)
```

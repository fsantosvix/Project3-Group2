```python
# Read data from csv file
street_lights_data = Path("../CSV files/Street_Lights.csv")
df = pd.read_csv(street_lights_data)
df.head()
```

```python
# List all columns
df.columns
```

```python
# Create a filtered version with only the relevant columns
reduced_df = df[['X', 'Y', 'LIGHTTYPE_DESC', 'NUMBERLIGHTS','STREETLIGHTID', 'STREETNAME', 'WATTAGE1']]

# Rename columns in the DataFrame to match SQL table column names
reduced_df = reduced_df.rename(columns={
    'X': 'X_coord',
    'Y': 'Y_coord',
    'LIGHTTYPE_DESC': 'light_type',
    'NUMBERLIGHTS': 'number_of_lights',
    'STREETLIGHTID': 'street_light_ID',
    'STREETNAME': 'street_name',
    'WATTAGE1': 'wattage1'
})

# Check the DataFrame with filtered and updated column names
reduced_df.head()
```

```python

# Calculate the total wattage and add as a new column
reduced_df["Total_Wattage"] = reduced_df['number_of_lights']*reduced_df['wattage1']
reduced_df.head()
```

```python
# Remove incomplete records
clean_df = reduced_df.dropna(how="any")
```

```python
# Export df as a csv file
clean_df.to_csv('../CSV files/filteredDF.csv', index=False)
```
